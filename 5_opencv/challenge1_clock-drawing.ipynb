```python
import cv2
import numpy as np
from math import cos, sin,pi
import math
x0,y0=256,256
r=256

def get_circle_ans(angle, length):
    global x0,y0,r
    x1 = x0 + (r - length) * cos(angle * pi / 180)
    y1 = y0 + (r - length) * sin(angle * pi /180)
    return int(x1),int(y1)

def draw_hand(img, point, hand):
    if hand == 'hour':
        print("draw hour ", point)
        cv2.line(img, (256, 256), point,(127,255,255), 15)
    elif hand == 'minute':
        print("draw minute ", point)
        cv2.line(img, (256, 256), point,(255,127,255), 8)
    elif hand == 'second':
        print("draw second ", point)
        cv2.line(img, (256, 256), point,(255,255,127), 2)
        
def time2angle(time, hand):
    if hand == 'hour':
        return (time-3)/12.0*360
    else:
        return (time-15)/60.0*360
    
import datetime
import time as th
img = np.zeros((512, 512, 3), np.uint8)
img[:] = (255,255,255)
img = cv2.circle(img, (256, 256), 250, (0, 0, 0), 5)

pt1 = []

# 3. 画出60条秒和分钟的刻线
for i in range(60):
    # 最外部圆，计算A点
    x1 = x0+(r-5)*math.cos(i*6*np.pi/180.0)
    y1 = y0+(r-5)*math.sin(i*6*np.pi/180.0)
    pt1.append((int(x1), int(y1)))

    # 同心小圆，计算B点
    x2 = x0+(r-15)*math.cos(i*6*np.pi/180.0)
    y2 = y0+(r-15)*math.sin(i*6*np.pi/180.0)

    cv2.line(img, pt1[i], (int(x2), int(y2)), (0, 0, 0), thickness=2)

# 4. 画出12条小时的刻线
for i in range(12):
    # 12条小时刻线应该更长一点
    x = x0+(r-25)*math.cos(i*30*np.pi/180.0)
    y = y0+(r-25)*math.sin(i*30*np.pi/180.0)
    # 这里用到了前面的pt1
    cv2.line(img, pt1[i*5], (int(x), int(y)), (0, 0, 0), thickness=5)
    x_mar = 0
    y_mar = 0
    hour_text = 0
    if i <= 9:
        hour_text = i + 3
    else:
        hour_text = i - 9
    if int(y) < 127 :
        y_mar = 30
    elif 127 <= int(y) <= 256:
        y_mar = 5
    elif int(y) > 256 :
        y_mar = -15
    if int(x) < 127 :
        x_mar = 5
    elif 127 <= int(x) <= 256:
        x_mar = -6
    elif int(x) > 256 :
        x_mar = -35
    cv2.putText(img,str(hour_text) ,(int(x)+x_mar, int(y)+y_mar),cv2.FONT_HERSHEY_SIMPLEX,1, (0,0,0), 2)
while True:
    temp = np.copy(img)
    # # cv2.namedWindow('image')
    now_time = datetime.datetime.now()
    time = datetime.datetime.now().strftime('%H:%M:%S').split(':')
    for i in range(len(time)):
        time[i] = int(time[i])
    
    # if int(time[0]) > 12:
    #     int(time[0]) = int(time[0]) - 12
    hour = time2angle(time[0], 'hour')
    minute = time2angle(time[1], 'minute')
    second = time2angle(time[2], 'second')
    
    hour_x,hour_y = get_circle_ans(hour, 65)
    minute_x,minute_y = get_circle_ans(minute, 35)
    second_x,second_y = get_circle_ans(second, 10)
    
    draw_hand(temp,(hour_x,hour_y),'hour')
    draw_hand(temp,(minute_x,minute_y),'minute')
    draw_hand(temp,(second_x,second_y),'second')
    print(second_x,second_y)
    cv2.imshow('circle', temp)
    th.sleep(1)
    if cv2.waitKey(1) == 27:  # 按下ESC键退出
        break
# cv2.line(img, (256, 256), (256,30),(0,0,0), 3)
# cv2.line(img, (256, 256), (50,256),(0,0,0), 5)
# cv2.line(img, (256, 256), (512,256),(0,0,0), 1)
# cv2.imshow('circle', img)
# 
# cv2.waitKey(0)

```
